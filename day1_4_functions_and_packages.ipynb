## Packages

A package is a collection of Python modules that provide specific functionalities. For example:
- **numpy** for numerical computations.
- **matplotlib** for plotting.
- **skimage** for image analysis.

To use a package, you first need to import it using the `import` statement. You can also install external packages using tools like `pip`.

Example: Using the `math` package to calculate the square root of a number.

````python
import math

# Use the sqrt function from the math package
print("Square root of 16:", math.sqrt(16))

# Use another math function
print("Value of pi:", math.pi)
````

## Example: Using a Package for Bioimage Analysis

The **skimage** (scikit-image) package is commonly used for processing and analyzing microscope images. Let's use it to:
1. Load a sample image.
2. Display the image.
3. Convert the image to grayscale.

````python
from skimage import data, color
import matplotlib.pyplot as plt

# Load a sample image
image = data.astronaut()

# Convert to grayscale
gray_image = color.rgb2gray(image)

# Display the original and grayscale images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(gray_image, cmap="gray")
ax[1].set_title("Grayscale Image")
ax[1].axis("off")

plt.tight_layout()
plt.show()
````
